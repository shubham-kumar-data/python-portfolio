# write a program to create electricity bill using user define function
if 500 units used - pay USD 5 for each unit
if 700 units used - pay USD 7 for each unit
if 1000 units used - pay USD 10 for each unit
if more than 1000 used  - Pay USD 15 for each unit

```python
def electricity(n):
    if n<=500:
         print( 'Your electricity bill is USD ', n*5)
    elif n>500 and n<= 700:
         print( 'Your electricity bill is USD ', n*7)
    elif n>700 and n<=1000:
        print( 'Your electricity bill is USD ', n*10)
    elif n> 1000:
         print( 'Your electricity bill is USD ', n*15)
```

```python
def instructions():
    print('The last date of paying bill is 20th July')
    print('If paid after 20th July an additional fine of USD 1000 will be charged')
    print('Electricity office: Arizona')
```

```python
for i in range(3):
    a = input( ' Please enter your name :')
    n = int(input('Please enter the units used :'))
    electricity(n)
    print()
    instructions()
    print()
```
