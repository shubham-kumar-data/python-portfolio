# Changing values in dic

```python
x = {

    "ID" : "123ab",
    "Color" : "Black",
    "year" : 1968    
}

x['year'] = 2026
print(x)
```

# Adding values in dic

```python
x = {

    "ID" : "123ab",
    "Color" : "Black",
    "year" : 1968    
}
print (x)
x.update({"country":"USA"})
print(x)
```

# Removing values in dic

```python
x = {

    "ID" : "123ab",
    "Color" : "Black",
    "year" : 1968, 
    "country":"USA"
}
print (x)
x.pop("year")
print(x)
```

# Deleting values in dic

```python
x = {

    "ID" : "123ab",
    "Color" : "Black",
    "year" : 1968, 
    "country":"USA"
}
print(x)
del x['year']
print (x)
```

# Empty the dict


```python
x = {

    "ID" : "123ab",
    "Color" : "Black",
    "year" : 1968, 
    "country":"USA"
}
x.clear()
print(x)
print(type(x))
```

# Concatination of Dict

```python
dict1 = {1:10,2:20,3:30}
dict2 = {3:30,4:40,5:50}
dict3 = {6:60,7:70,8:80}
dict4= {}
for d in (dict1,dict2,dict3): dict4.update(d)
print(dict4)
```
