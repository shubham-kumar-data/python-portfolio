# Adding items in the set

```python
a = {"x","y","z"}
print(a)
a.add(100)
print(a)
```

```python
a = {"data","python","code"}
b = {"x","y","z"}
a.update(b)
print(a)
```

# Removing set items 

```python
a = {"data","python","code"}
print(a)
a.remove("data")
print(a)
```

```python
a = {"data","python","code"}
print(a)
a.clear()
print(a)
```

# Joining two sets

```python
x = {"data","python","code"}
a = {"x","y","z"}
z = x.union(a)
print(z)
```
