# part 2 Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['English', 'Hindi' ,'Science', 'Maths'] 
plt.pie(x, labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ['English', 'Hindi' ,'Science', 'Maths'] 
c = ['Yellow', 'Magenta', 'brown','aqua']
plt.pie(x, labels = y, colors = c)
plt.show()
```

```python
x = [10,20,30,40]
y = ['English', 'Hindi' ,'Science', 'Maths'] 
c = ['Yellow', 'Magenta', 'brown','aqua']
plt.pie(x, labels = y, colors = c)
plt.legend()
plt.show()
```
