# Seaborn Graph

```python
var  = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
```

```python
sns.lineplot(x = var, y =var1)
plt.show()
```

```python
var  = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
import pandas as pd

df = pd.DataFrame({
    'var' :var,
    'var1': var1
})
sns.lineplot(x= 'var',y = 'var1', data = df)
```

```python
sns.lineplot(x = 'bill_depth_mm', y = 'flipper_length_mm', data = df1, hue = 'sex', style='sex',palette = 'rocket_r', markers = ["o",">"])
plt.grid()
plt.show()
```

```python
sns.lineplot(x = 'bill_depth_mm', y = 'flipper_length_mm', data = df1, hue = 'sex', style='sex',palette = 'rocket_r', markers = ["o",">"])
plt.grid()
plt.title('This is a line graph realted to penguins')
plt.show()
```
