# Seaborn Part 2 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('penguins')
```

```python
df
```

```python
df.describe()
```

df[['island', 'species']]

```python
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex', data = df)
plt.grid()
plt.show()
```

# order placement 

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df)
plt.grid()
plt.show()
```

# Hue Placement 

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df, hue_order = ['Female', 'Male'])
plt.grid()
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df, hue_order = ['Female', 'Male'], palette = 'Accent')
plt.grid()
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df, hue_order = ['Female', 'Male'], palette = 'gist_earth')
plt.grid()
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df, hue_order = ['Female', 'Male'], palette = 'gist_heat_r')
plt.grid()
plt.show()
```

# Saturation 

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', hue = 'sex',order = order1, data = df, hue_order = ['Female', 'Male'],
            palette = 'gist_heat_r', saturation = 0.2)
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Bins 

```python
sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240])
plt.show()
```

# color

```python
sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240], color = 'g')
plt.show()
```

```python
sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240], color = 'g', kde = True)
plt.show()
```

# kde

```python
sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240], color = 'r', kde = True, color = "green")
plt.show()
```
