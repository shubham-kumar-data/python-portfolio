# adding titles 

```python
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

fig = go.Figure(
    data=go.Scatter(x=x, y=y),
    layout=go.Layout(
        title='Line Graph',
        xaxis_title='X Axis',
        yaxis_title='Y Axis'
    )
)
fig.show()
```

```python
import plotly.graph_objects as go

# Define data
x1, y1 = [1,2,3,4,5], [5,4,6,7,3]
x2, y2 = [1,2,3,4,5], [4,3,6,2,1]
x3, y3 = [1,2,3,4,5], [8,6,8,2,6]  

# Create figure
fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=x1, y=y1, name='Line 1'))
fig.add_trace(go.Scatter(x=x2, y=y2, name='Line 2'))
fig.add_trace(go.Scatter(x=x3, y=y3, name='Line 3'))

# Show
fig.show()
```

```python
fig.add_trace(go.Scatter(x=x1, y=y1, name="line1", mode="lines", line=dict(color='red')))
fig.add_trace(go.Scatter(x=x2, y=y2, name="line2", mode="markers", marker=dict(color='blue', size=10)))
fig.add_trace(go.Scatter(x=x3, y=y3, name="line3", mode="lines+markers", line=dict(color='green'), marker=dict(color='green', size=8)))
```

# Bubble charts


```python
import plotly.graph_objects as go

x = [12, 3, 4, 5]
y = [4, 6, 8, 10]

fig = go.Figure(go.Scatter(x=x, y=y, mode="markers"))
fig.show()
```

```python
fig = go.Figure(go.Scatter(x=x, y=y, mode="markers", marker_size = [40,60,70,80]))
fig.show()
```
