# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7,2]
y = [10,5,8,9,4,4]
plt.plot(x,y, marker = 'o')
plt.show()
```

```python
x = [5,2,9,4,7,2]
y = [10,5,8,9,4,4]

plt.plot(x,y)
plt.plot(y, marker = '>')
plt.show()
```

```python
x = [5,2,9,4,7,2]
y = [10,5,8,9,4,4]

plt.plot(x,y)
plt.plot(y, marker = '>')
plt.plot(x, marker = 'o')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = ['r','g','b','y',]
plt.bar(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = ['r','g','b','y',]
plt.barh(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = ['r','g','b','y',]
plt.bar(x,y, color = c, width = 1.5)
plt.show()
```

# Scatter Plot

```python

x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.xlabel('month')
plt.ylabel('Number')
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.xlabel('month')
plt.ylabel('Number')
plt.scatter(x,y,color =  'r')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.xlabel('month')
plt.ylabel('Number')
plt.scatter(x,y,color =  'r', s = 350)
plt.show()
```
